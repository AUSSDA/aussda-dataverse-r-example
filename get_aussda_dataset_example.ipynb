```python
from pyDataverse.api import NativeApi, DataAccessApi
from pyDataverse.models import Dataverse
```

```python
base_url = 'https://data.aussda.at/'
api = NativeApi(base_url)
data_api = DataAccessApi(base_url)
```

```python
# Data citation: Kittel, Bernhard; Kritzinger, Sylvia; Boomgaarden, Hajo; Prainsack, Barbara; Eberl, Jakob-Moritz; Kalleitner, Fabian; Lebernegg, Noëlle S.; Partheymüller, Julia; Plescia, Carolina; Schiestl, David W.; Schlogl, Lukas, 2020, "Austrian Corona Panel Project (OA edition)", https://doi.org/10.11587/P5YJ0O, AUSSDA, V2, UNF:6:WBG5D3c7z25zyzqvlLvWkQ== [fileUNF]
# https://doi.org/10.11587/P5YJ0O
DOI = "doi:10.11587/P5YJ0O"
dataset = api.get_dataset(DOI)
```

```python
files_list = dataset.json()['data']['latestVersion']['files']
```

```python
for file in files_list:
     filename = file["dataFile"]["filename"]
     file_id = file["dataFile"]["id"]
     print("File name {}, id {}".format(filename, file_id))
```
